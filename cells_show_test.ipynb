```python
def get_commit_id(repo_root, branch='origin/master'):
    command = "git -C " + repo_root + " rev-parse " + branch
    try:
        output = subprocess.check_output(command.split(" "), stderr=subprocess.STDOUT).decode("utf-8").strip()
    except subprocess.CalledProcessError as e:
        warnings.warn("Warning: Failed to get commit from branch " + branch + "\n" + str(e.output.decode("utf-8")))
        raise

    return output

def get_repo_root(file_dir):
    command = "git -C " + file_dir + " rev-parse --show-toplevel"

    try:
        output = subprocess.check_output(command.split(" "), stderr=subprocess.STDOUT).decode("utf-8").strip()
    except subprocess.CalledProcessError as e:
        print(str(e.output.decode("utf-8")))
        raise
        
    return output
        
        
def is_synced(file_dir):

    try:
        repo_root = get_repo_root(file_dir)
        commit1 = get_commit_id(repo_root, "masterdd")
        commit2 = get_commit_id(repo_root, "origin/master")
    except subprocess.CalledProcessError as e:
        raise

    if commit1 != commit2:   
        return False

    return True
```
